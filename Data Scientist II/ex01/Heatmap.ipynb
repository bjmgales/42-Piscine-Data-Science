```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
from matplotlib.colors import LinearSegmentedColormap
import pandas as pd
import numpy as np
df = pd.read_csv('../../subject/Knights/Train_Knight.csv')
```

```python
df['knight'] = df['knight'].map({'Jedi': 0, 'Sith': 1})
df = df.corr()
fig, ax = plt.subplots()
ax.set_xticks(range(len(df.columns)))
ax.set_xticklabels(df.columns, rotation=90)
ax.set_yticks(range(len(df.columns)))
ax.set_yticklabels(df.columns)
colors=['#04051A', '#3A1A41', '#782059', '#B81657', '#EB473E',
        '#F4835C', '#F8D1B8', '#FAEBDD']
custom_cmap = LinearSegmentedColormap.from_list("", colors, N=256)
cax = ax.matshow(df, cmap=custom_cmap)
ax.yaxis.set_major_locator(MultipleLocator(2))  # Draw ticks + label every two data points
ax.xaxis.set_major_locator(MultipleLocator(2))
fig.colorbar(cax)
plt.show()
```
