```python
import matplotlib.pyplot as plt
import pandas as pd

df_train = pd.read_csv('../../subject/Knights/Train_Knight.csv')
df_test = pd.read_csv('../../subject/Knights/Test_Knight.csv')
```

```python
figTrain = plt.figure(figsize=(13, 5))
name = ['Awareness', 'Stims', 'left']
for i in range(2):
    ax = plt.subplot2grid([1, 2], (0, i), fig=figTrain)
    plt.scatter(df_train[df_train['knight']=='Jedi'][name[0]], df_train[df_train['knight']=='Jedi'][name[1]],
                color='blue', alpha=0.4, label='Jedi')
    plt.scatter(df_train[df_train['knight']=='Sith'][name[0]], df_train[df_train['knight']=='Sith'][name[1]],
                color='red', alpha=0.4, label='Sith')
    plt.xlabel(name[0])
    plt.ylabel(name[1])
    ax.legend(loc=f"upper {name[2]}")
    name = ['Push', 'Deflection', 'right']
plt.show()

figTest = plt.figure(figsize=(13, 5))
name = ['Awareness', 'Stims', 'left']
for i in range(2):
    ax = plt.subplot2grid([1, 2], (0, i), fig=figTest)
    plt.scatter(df_test[name[0]], df_test[name[1]], color='green', alpha=0.4, label='Knight')
    plt.xlabel(name[0])
    plt.ylabel(name[1])
    ax.legend(loc=f"upper {name[2]}")
    name = ['Deflection', 'Push', 'right']

plt.show()
```
