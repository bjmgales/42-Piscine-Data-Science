```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('../../subject/Knights/Train_Knight.csv')
```

```python
scaler = StandardScaler()
if 'knight' in df.columns:
    defined_knight = True
    type_save = df['knight']
    df = df.drop('knight', axis=1)
else :
    defined_knight = False

print(df.head(3).to_string(index=False))

original_columns = df.columns
df = pd.DataFrame(scaler.fit_transform(df), columns=original_columns)

if defined_knight is True:
    df['knight'] = type_save
    plt.scatter(df[df['knight']=='Jedi']['Awareness'], df[df['knight']=='Jedi']['Stims'],
                color='blue', alpha=0.4, label='Jedi')
    plt.scatter(df[df['knight']=='Sith']['Awareness'], df[df['knight']=='Sith']['Stims'],
                color='red', alpha=0.4, label='Sith')
else:
    plt.scatter(df['Awareness'], df['Stims'],
                color='green', alpha=0.4, label='Knight')
print(df.head(3).to_string(index=False))


plt.xlabel('Awareness')
plt.ylabel('Stims')
plt.legend(loc="upper left")

plt.show()
```
