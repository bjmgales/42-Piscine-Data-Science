```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('../../subject/Knights/Test_Knight.csv')
```

<div style="border: 2px solid white; padding: 10px;">

Histogram Explanation:
- X-axis: Represents the note/value of the skill.
- Y-axis: Represents the number of Knights with that skill value.
    
        Example: If 10 Knights have an Agility Skill of 5, the bar at x = 5 will have a height of 10.

Understanding Bins:
    The bins define the width of the bars in the histogram:

- Fewer bins:
    - Each bin covers a larger range of skill values.
    - More Knights will fall into the same bar, resulting in less precise data.
- More bins:
    - Each bin covers a smaller range of skill values.
    - The histogram becomes more precise, but each bar represents fewer Knights.
</div>

```python
fig = plt.figure(figsize=(40, 30))
name_index = 0
for i in range(6):
    for j in range(5):
        ax = plt.subplot2grid([6, 5], (i, j), fig=fig)
        ax.hist(df[df.columns[name_index]], bins=40, color='#7fbf7f', label='Knight')
        ax.set_title(df.columns[name_index])
        ax.legend(loc="upper right")
        name_index += 1
plt.show()
```

```python
df = pd.read_csv('../../subject/Knights/Train_Knight.csv')

fig = plt.figure(figsize=(40, 30))
name_index = 0
for i in range(6):
    for j in range(5):
        ax = plt.subplot2grid([6, 5], (i, j), fig=fig)
        sith_data = df[df['knight'] == 'Sith'][df.columns[name_index]]
        jedi_data = df[df['knight'] == 'Jedi'][df.columns[name_index]]
        ax.hist(sith_data, bins=40, color='salmon', alpha=0.5, label='Sith')
        ax.hist(jedi_data, bins=40, color='blue', alpha=0.5, label='Jedi')
        ax.set_title(df.columns[name_index])
        ax.legend(loc="upper right")
        name_index += 1
plt.show()
```
